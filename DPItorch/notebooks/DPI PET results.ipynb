```python
# --- 用户配置区域 ---
# 请根据您的实际情况修改以下路径和参数

import os

def get_model_and_loss_paths(jobId, base_dir="../checkpoint"):
    """
    根据 jobId 自动在对应目录中查找唯一的 .pth 和 .npy 文件。
    
    Parameters:
        jobId (int or str): 任务编号。
        base_dir (str): checkpoint 根目录路径，默认 "../checkpoint"。
    
    Returns:
        model_path (str): 找到的 .pth 文件路径。
        loss_path (str): 找到的 .npy 文件路径。
    
    Raises:
        FileNotFoundError: 如果文件不存在或找到多个匹配文件。
    """
    jobId = str(jobId)
    dir_path = os.path.join(base_dir, f"train{jobId}")
    
    if not os.path.isdir(dir_path):
        raise FileNotFoundError(f"目录不存在: {dir_path}")
    
    # 查找唯一的 .pth 文件
    pth_files = [f for f in os.listdir(dir_path) if f.endswith(".pth")]
    if len(pth_files) != 1:
        raise FileNotFoundError(f"{dir_path} 中找到 {len(pth_files)} 个 .pth 文件，应为 1 个。")
    
    # 查找唯一的 .npy 文件
    npy_files = [f for f in os.listdir(dir_path) if f.endswith(".npy")]
    if len(npy_files) != 1:
        raise FileNotFoundError(f"{dir_path} 中找到 {len(npy_files)} 个 .npy 文件，应为 1 个。")

    model_path = os.path.join(dir_path, pth_files[0])
    loss_path = os.path.join(dir_path, npy_files[0])
    
    return model_path, loss_path


# MODEL_PATH = "../checkpoint/train641601/realnvp_flow16_ld1e+00_tv1e+00_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = "../checkpoint/train641601/loss_realnvp_flow16_ld1e+00_tv1e+00_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径
# MODEL_PATH = "../checkpoint/train643200/realnvp_flow32_ld1e+00_tv1e+00_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = "../checkpoint/train643200/loss_realnvp_flow32_ld1e+00_tv1e+00_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径
# jobId = 641602
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/realnvp_flow16_ld1e+00_tv1e+03_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_realnvp_flow16_ld1e+00_tv1e+03_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径
# jobId = 641603
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/realnvp_flow16_ld1e+00_tv1e+02_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_realnvp_flow16_ld1e+00_tv1e+02_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径


# jobId = 641604
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/realnvp_flow16_ld1e+00_tv1e+01_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_realnvp_flow16_ld1e+00_tv1e+01_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径
# jobId = 641607
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/realnvp_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_realnvp_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径

# jobId = 641608
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/glow_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_glow_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径

# jobId = 641609
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/glow_flow16_ld1e+00_tv1e+01_l10e+00_lr5e-05_batch64_size64_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_glow_flow16_ld1e+00_tv1e+01_l10e+00_lr5e-05_batch64_size64_adaptScale.npy" # 修改为您的损失 .npy 文件路径

# jobId = 1281600
# DIR = f"../checkpoint/train{jobId}"
# MODEL_PATH = f"{DIR}/glow_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch32_size128_adaptScale_final.pth"  # 修改为您的模型 .pth 文件路径
# LOSS_CURVE_PATH = f"{DIR}/loss_glow_flow16_ld1e+00_tv5e+00_l10e+00_lr5e-05_batch32_size128_adaptScale.npy" # 修改为您的损失 .npy 文件路径

jobId = 1281603
MODEL_PATH, LOSS_CURVE_PATH = get_model_and_loss_paths(jobId)

print("Model Path:", MODEL_PATH)
print("Loss Curve Path:", LOSS_CURVE_PATH)


# 数据路径 (与训练时使用的数据对应)
# 您需要根据训练时使用的 image_size 来设置这些路径
IMAGE_SIZE = jobId//10000 # 或 128，根据您的模型
print(f'IMAGE_SIZE = {IMAGE_SIZE}*{IMAGE_SIZE}')

FRAME_IDX = 10  # 与训练时相同
ROI_IDX = 30    # 与训练时相同

# 生成模型参数 (应与训练时模型结构一致)
N_FLOW = 16 # 例如，或者您在训练时使用的值
MODEL_FORM = "glow" # 或 "glow"
N_BLOCK = 4 # for Glow
```

```python

# --- 2. 加载训练好的模型 ---
print(f"加载模型: {MODEL_PATH}")
# 重新导入模型定义 (确保与训练脚本中的定义一致)
# 您可能需要将 generative_model.py 或其部分内容复制到此 notebook 或确保它在 Python 路径中
# 例如: from generative_model import realnvpfc_model, glow_model (假设这些在您的环境中可用)
# 这里我们用 DPI_PET.py 中的定义作为参考
# (确保 RealNVP 和 ImgLogScale 类定义在此处可用或已导入)
# ---- 开始：从DPI_PET.py复制的模型和辅助类定义 ----
class ImgLogScale(torch.nn.Module):
    def __init__(self, init_scale=1.0):
        super().__init__()
        self.log_scale = torch.nn.Parameter(torch.tensor([np.log(init_scale)], dtype=torch.float32))
    def forward(self):
        return self.log_scale

# 假设 realnvpfc_model 和 glow_model 可以在环境中找到
# 例如，如果它们在 DPItorch 目录下，可以这样添加路径：
import sys
sys.path.append(str(Path.cwd().parent)) # 添加 DPItorch 目录到 sys.path
from generative_model import realnvpfc_model, glow_model

N_PIX = IMAGE_SIZE * IMAGE_SIZE
if MODEL_FORM == 'realnvp':
    Gnet = realnvpfc_model.RealNVP(N_PIX, N_FLOW, affine=True).to(device)
    print("RealNVP 模型已构建")
elif MODEL_FORM == 'glow':
    z_shapes = glow_model.calc_z_shapes(1, IMAGE_SIZE, N_FLOW, N_BLOCK) # 假设 n_channel=1
    Gnet = glow_model.Glow(1, N_FLOW, N_BLOCK, True, True).to(device) # 假设 n_channel=1
    print("Glow 模型已构建")
else:
    raise ValueError(f"不支持的模型类型: {MODEL_FORM}")

checkpoint = torch.load(MODEL_PATH, map_location=device)
Gnet.load_state_dict(checkpoint['model_state_dict'])
Gnet.eval() # 设置为评估模式
print("模型权重已加载.")

use_learnable_scale = 'logscale_state_dict' in checkpoint
if use_learnable_scale:
    logscale_module = ImgLogScale(init_scale=1.0).to(device) # init_scale不重要，因为会加载
    logscale_module.load_state_dict(checkpoint['logscale_state_dict'])
    logscale_module.eval()
    learned_scale_factor = torch.exp(logscale_module()).item()
    print(f"可学习缩放因子已加载，值为: {learned_scale_factor:.4f}")
else:
    # 检查是否保存了固定缩放的信息（DPI_PET.py的较新版本会保存）
    if 'fixed_scale' in checkpoint:
        fixed_scale_factor = checkpoint['fixed_scale']
        print(f"模型训练时使用了固定缩放或无显式缩放，参考固定缩放值: {fixed_scale_factor:.4f}")
    else: # 旧模型可能没有这个字段
        fixed_scale_factor = np.mean(act_gt) # Fallback or a placeholder
        print(f"未找到固定缩放因子信息，使用act_gt均值作为参考: {fixed_scale_factor:.4f}")
    learned_scale_factor = None # 表示不使用可学习的
```

```python

import numpy as np

def SSIM(img1: np.ndarray, img2: np.ndarray) -> float:
    """
    Compute global SSIM between two normalized images using vector formulation.
    Args:
        img1: Ground truth image, shape (H, W), values in [0, 1]
        img2: Generated image, shape (H, W), values in [0, 1]
    Returns:
        SSIM scalar
    """

    # Flatten images
    img1 = img1.astype(np.float32).flatten()
    img2 = img2.astype(np.float32).flatten()

    # Constants
    K1 = 0.01
    K2 = 0.03
    L = 1.0
    C1 = (K1 * L) ** 2
    C2 = (K2 * L) ** 2
    C3 = C2 / 2

    # Means and standard deviations
    mu1 = np.mean(img1)
    mu2 = np.mean(img2)
    sigma1 = np.std(img1)
    sigma2 = np.std(img2)
    sigma12 = np.mean((img1 - mu1) * (img2 - mu2))

    # SSIM components
    lxy = (2 * mu1 * mu2 + C1) / (mu1**2 + mu2**2 + C1)
    cxy = (2 * sigma1 * sigma2 + C2) / (sigma1**2 + sigma2**2 + C2)
    sxy = (sigma12 + C3) / (sigma1 * sigma2 + C3)

    ssim = lxy * cxy * sxy
    return ssim

# --- 5. 定量评估 - 指标计算 ---
# print("计算定量评估指标...")
# # 打印输入图像的当前范围，以确认它们是否为原始尺度
# print(f"输入图像范围检查 - GT: [{act_gt.min():.4f}, {act_gt.max():.4f}]")
# print(f"输入图像范围检查 - Mean Recon: [{mean_recon_img.min():.4f}, {mean_recon_img.max():.4f}]")
# print(f"输入图像范围检查 - OSEM: [{osem_res.min():.4f}, {osem_res.max():.4f}]")
# print(f"输入图像范围检查 - FBP: [{fbp_res.min():.4f}, {fbp_res.max():.4f}]")
# if 'all_samples_np' in locals() and len(all_samples_np) > 0: # 确保 all_samples_np 已定义且非空
#     print(f"输入图像范围检查 - Sample 0: [{all_samples_np[0].min():.4f}, {all_samples_np[0].max():.4f}]")
# else:
#     print("警告: 'all_samples_np' 未定义或为空，无法检查其范围。")


mse_val = mean_squared_error(act_gt, mean_recon_img)
nrmse_val = normalized_root_mse(act_gt, mean_recon_img)
psnr_val = peak_signal_noise_ratio(act_gt, mean_recon_img, data_range=act_gt.max()-act_gt.min())
ssim_val = structural_similarity(act_gt, mean_recon_img, data_range=act_gt.max()-act_gt.min())
ssim_val2 = SSIM(act_gt, mean_recon_img)

print(f"ML Recon:")
# print(f"  MSE : {mse_val:.6f}")
print(f"  NRMSE : {nrmse_val:.6f}")
print(f"  PSNR: {psnr_val:.2f} dB")
# print(f"  SSIM: {ssim_val:.4f}")
print(f"  SSIM: {ssim_val2:.4f}")


osem_mse_val = mean_squared_error(act_gt, osem_res)
osem_nrmse_val = normalized_root_mse(act_gt, osem_res)
osem_psnr_val = peak_signal_noise_ratio(act_gt, osem_res, data_range=act_gt.max()-act_gt.min())
osem_ssim_val = structural_similarity(act_gt, osem_res, data_range=act_gt.max()-act_gt.min())
osem_ssim_val2 = SSIM(act_gt, osem_res)


print(f"OSEM Recon:")
# print(f"  MSE : {osem_mse_val:.6f}")
print(f"  NRMSE : {osem_nrmse_val:.6f}")
print(f"  PSNR: {osem_psnr_val:.2f} dB")
# print(f"  SSIM: {osem_ssim_val:.4f}")
print(f"  SSIM: {osem_ssim_val2:.4f}")


fbp_mse_val = mean_squared_error(act_gt, fbp_res)
fbp_nrmse_val = normalized_root_mse(act_gt, fbp_res)
fbp_psnr_val = peak_signal_noise_ratio(act_gt, fbp_res, data_range=act_gt.max()-act_gt.min())
fbp_ssim_val = structural_similarity(act_gt, fbp_res, data_range=act_gt.max()-act_gt.min())
fbp_ssim_val2 = SSIM(act_gt, fbp_res)


print(f"FBP Recon:")
# print(f"  MSE : {fbp_mse_val:.6f}")
print(f"  NRMSE : {fbp_nrmse_val:.6f}")
print(f"  PSNR: {fbp_psnr_val:.2f} dB")
# print(f"  SSIM: {fbp_ssim_val:.4f}")
print(f"  SSIM: {fbp_ssim_val2:.4f}")

# 您也可以计算每个样本的指标，然后报告平均值和标准差
# mses_samples = [mean_squared_error(act_gt, all_samples_np[i]) for i in range(N_SAMPLES)]
# nrmses_samples = [normalized_root_mse(act_gt, all_samples_np[i]) for i in range(N_SAMPLES)]
# psnrs_samples = [peak_signal_noise_ratio(act_gt, all_samples_np[i], data_range=act_gt.max()-act_gt.min()) for i in range(N_SAMPLES)]
# ssims_samples = [structural_similarity(act_gt, all_samples_np[i], data_range=act_gt.max()-act_gt.min()) for i in range(N_SAMPLES)]

# print(f"\n对于 {N_SAMPLES} 个独立样本 (均值 ± 标准差):")
# print(f"  MSE : {np.mean(mses_samples):.6f} ± {np.std(mses_samples):.6f}")
# print(f"  NRMSE : {np.mean(nrmses_samples):.6f} ± {np.std(nrmses_samples):.6f}")
# print(f"  PSNR: {np.mean(psnrs_samples):.2f} ± {np.std(psnrs_samples):.2f} dB")
# print(f"  SSIM: {np.mean(ssims_samples):.4f} ± {np.std(ssims_samples):.4f}")

# 输出对比表
print("\n=== 图像重建指标对比 ===")
print(f"{'Method':<13} | {'NRMSE':<8} | {'PSNR':<8} | {'SSIM':<8}")
print("-" * 50)
print(f"{'ML Recon':<13} | {nrmse_val:.6f} | {psnr_val:.2f} dB | {ssim_val2:.4f}")
print(f"{'OSEM Recon':<13} | {osem_nrmse_val:.6f} | {osem_psnr_val:.2f} dB | {osem_ssim_val2:.4f}")
print(f"{'FBP Recon':<13} | {fbp_nrmse_val:.6f} | {fbp_psnr_val:.2f} dB | {fbp_ssim_val2:.4f}")
print("-" * 50)

```
